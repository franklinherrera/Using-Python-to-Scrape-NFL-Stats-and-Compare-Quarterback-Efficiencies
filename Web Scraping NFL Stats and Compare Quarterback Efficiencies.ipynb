<h2>Data Wrangling</h2>

<h3>Gathering and Loading Data</h3>

Data source: https://www.pro-football-reference.com/years/2019/passing.htm

Interedted Statistical categories :
- Completion percentage — Cmp%
- Passing yards — Yds
- Passing touchdowns — TD
- Interceptions — Int
- Yards per attempt — Y/A
- Passer rating — Rate

```python
#Import scraping modules
from urllib.request import urlopen
from bs4 import BeautifulSoup

#Import data manipulation modules
import pandas as pd
import numpy as np

#Import data visualization modules
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
```

### Scrape Data

```python
#Page URL
url = 'https://www.pro-football-reference.com/years/2019/passing.htm'

#open url and pass to BeautifulSoup
html = urlopen(url)
stats_page = BeautifulSoup(html)
```

```python
#Collecting table headers
column_headers = stats_page.findAll('tr')[0]
column_headers = [i.getText() for i in column_headers.findAll('th')]
print(column_headers)
```

```python
#Collecting table rows
rows = stats_page.findAll('tr')[1:]

#Getting stats from each row
qb_stats = []
for i in range(len(rows)):
    qb_stats.append([col.getText() for col in rows[i].findAll('td')])
    
print(qb_stats[0])
```

```python
#Creating a Dataframe from our scraped data
df = pd.DataFrame(qb_stats, columns=column_headers[1:])
df.head()
```

<h3>Assessing and Cleaning Data</h3>

```python
df.columns
```

- There are two columns titled `Yds`; one is for passing yards, and the other is yards lost due to being sacked.

```python
new_columns = df.columns.values
new_columns[-6] = 'Yds_Sack'
df.columns = new_columns
```

```python
df.columns
```

```python
# Select stat categories
categories = ['Cmp%', 'Yds', 'TD', 'Int', 'Y/A', 'Rate']
```

```python
# Creating a subset dataframe for radar chart
df_radar = df[['Player', 'Tm'] + categories]
df_radar.head()
```

```python
# Checking data types
df_radar.dtypes
```

```python
# Converting data to numerical values
for i in categories:
    df_radar[i] = pd.to_numeric(df[i])
```

```python
# Test: Checking data types
df_radar.dtypes
```

```python
# Removing ornamental characters for achievements
df_radar['Player'] = df_radar['Player'].str.replace('*','')
df_radar['Player'] = df_radar['Player'].str.replace('+','')
```

```python
# Filtering quarterbacks who threw more than 1500 yards
df_radar_filtered = df_radar.query('Yds > 1500')
df_radar_filtered.head()
```

```python
# Creating columns with percentile rank
for i in categories:
    df_radar_filtered[i + '_Rank'] = df_radar_filtered[i].rank(pct=True)
    
# We need to flip the rank for interceptions
df_radar_filtered['Int_Rank'] = 1 - df_radar_filtered['Int_Rank']

df_radar_filtered.head()
```

<h3>Generating Radar Charts</h3>

```python
# General plot parameters
mpl.rcParams['font.family'] = 'Avenir'
mpl.rcParams['font.size'] = 16
mpl.rcParams['axes.linewidth'] = 0
mpl.rcParams['xtick.major.pad'] = 15
```

```python
# Setting Teams color
team_colors = {'ARI':'#97233f', 'ATL':'#a71930', 'BAL':'#241773', 'BUF':'#00338d', 'CAR':'#0085ca', 'CHI':'#0b162a', 'CIN':'#fb4f14', 'CLE':'#311d00', 'DAL':'#041e42', 'DEN':'#002244', 'DET':'#0076b6', 'GNB':'#203731', 'HOU':'#03202f', 'IND':'#002c5f', 'JAX':'#006778', 'KAN':'#e31837', 'LAC':'#002a5e', 'LAR':'#003594', 'MIA':'#008e97', 'MIN':'#4f2683', 'NWE':'#002244', 'NOR':'#d3bc8d', 'NYG':'#0b2265', 'NYJ':'#125740', 'OAK':'#000000', 'PHI':'#004c54', 'PIT':'#ffb612', 'SFO':'#aa0000', 'SEA':'#002244', 'TAM':'#d50a0a', 'TEN':'#0c2340', 'WAS':'#773141'}
```

"The angles at which we plot all our points will be dependent on the number of statistical categories we have. In our example, we have 6 categories, so we will plot our points every 2π/6 radians, or 60 degrees. We can calculate this as follows, using numpy.linspace() :"

```python
# Calculating angles for radar chart
offset = np.pi/6
angles = np.linspace(0, 2*np.pi, len(categories) + 1) + offset
```

"We are adding the offset term to adjust where the labels appear on the circle (instead of the first label at 0 radians, it now appears at π/6)."

```python
# Function to create radar chart
def create_radar_chart(ax, angles, player_data, color='blue'):
    
    # Plot data and fill with team color
    ax.plot(angles, np.append(player_data[-(len(angles)-1):], 
            player_data[-(len(angles)-1)]), color=color, linewidth=2)
    ax.fill(angles, np.append(player_data[-(len(angles)-1):], 
            player_data[-(len(angles)-1)]), color=color, alpha=0.2)
    
    # Set category labels
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(categories)
    
    # Remove radial labels
    ax.set_yticklabels([])

    # Add player name
    ax.text(np.pi/2, 1.7, player_data[0], ha='center', va='center', 
            size=18, color=color)
    
    # Use white grid
    ax.grid(color='white', linewidth=1.5)

    # Set axis limits
    ax.set(xlim=(0, 2*np.pi), ylim=(0, 1))

    return ax
```

```python
# Function to get QB data
def get_qb_data(data, team):
  return np.asarray(data[data['Tm'] == team])[0]
```

<h3>Visualizing Data</h3>

#### NFC West

```python
# Creating figure
fig = plt.figure(figsize=(8,8), facecolor='white')

# Adding subplots
ax1 = fig.add_subplot(221, projection='polar', facecolor='#ededed')
ax2 = fig.add_subplot(222, projection='polar', facecolor='#ededed')
ax3 = fig.add_subplot(223, projection='polar', facecolor='#ededed')
ax4 = fig.add_subplot(224, projection='polar', facecolor='#ededed')

# Adjusting space between subplots
plt.subplots_adjust(hspace=0.8, wspace=0.5)

# Getting QB Data
sf_data = get_qb_data(df_radar_filtered, 'SFO')
sea_data = get_qb_data(df_radar_filtered, 'SEA')
ari_data = get_qb_data(df_radar_filtered, 'ARI')
lar_data = get_qb_data(df_radar_filtered, 'LAR')

# Plotting QB Data
ax1 = create_radar_chart(ax1, angles, lar_data, team_colors['LAR'])
ax2 = create_radar_chart(ax2, angles, ari_data, team_colors['ARI'])
ax3 = create_radar_chart(ax3, angles, sea_data, team_colors['SEA'])
ax4 = create_radar_chart(ax4, angles, sf_data, team_colors['SFO'])

plt.show();
```

- Russell Wilson is on an elite level, reaching at least 75th percentiles in all statistical categories. This analysis also doesn’t take into account what he brings with his scrambling ability.

- Jimmy Garoppolo is generally a solid QB, but throws a few boneheaded interceptions from time-to-time.

- Jared Goff gets a lot of yards, but it does not seem to translate to a lot of production (i.e. “empty yards”).

- Kyler Murray is hard to judge because he was a rookie, but his efficiency looks to be the lowest of the four. Like Wilson, his mobility is not accounted for in this plot.

#### MVP Race

```python
# MVP Race
# Creating figure
fig = plt.figure(figsize=(8, 4), facecolor='white')

# Adding subplots
ax1 = fig.add_subplot(121, projection='polar', facecolor='#ededed')
ax2 = fig.add_subplot(122, projection='polar', facecolor='#ededed')

# Adjusting space between subplots
plt.subplots_adjust(hspace=0.8, wspace=0.5)

# Getting QB data
bal_data = get_qb_data(df_radar_filtered, 'BAL')
sea_data = get_qb_data(df_radar_filtered, 'SEA')

# Plotting QB data
ax1 = create_radar_chart(ax1, angles, sea_data, team_colors['SEA'])
ax2 = create_radar_chart(ax2, angles, bal_data, team_colors['BAL'])

plt.show();
```

From a pure passer standpoint, Russell Wilson has more balanced stats, but Lamar Jackson clearly had the more “bang-for-buck” passing yards, as shown by his number of passing TDs and quarterback rating. Additionally, Lamar Jackson set the QB rushing record for a season, which is not captured in this chart.

```python
# 1st Round Draft Picks
# Creating figure
fig = plt.figure(figsize=(8, 8), facecolor='white')

# Adding subplots
ax1 = fig.add_subplot(221, projection='polar', facecolor='#ededed')
ax2 = fig.add_subplot(222, projection='polar', facecolor='#ededed')
ax3 = fig.add_subplot(223, projection='polar', facecolor='#ededed')
ax4 = fig.add_subplot(224, projection='polar', facecolor='#ededed')

# Adjusting space between subplots
plt.subplots_adjust(hspace=0.8, wspace=0.5)

# Getting QB data
cin_data = get_qb_data(df_radar_filtered, 'CIN')
mia_data = get_qb_data(df_radar_filtered, 'MIA')
lac_data = get_qb_data(df_radar_filtered, 'LAC')
gnb_data = get_qb_data(df_radar_filtered, 'GNB')

# Plotting QB data
ax1 = create_radar_chart(ax1, angles, cin_data, team_colors['CIN'])
ax2 = create_radar_chart(ax2, angles, mia_data, team_colors['MIA'])
ax3 = create_radar_chart(ax3, angles, lac_data, team_colors['LAC'])
ax4 = create_radar_chart(ax4, angles, gnb_data, team_colors['GNB'])

plt.show();
```
